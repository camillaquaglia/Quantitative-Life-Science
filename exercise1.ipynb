```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
k = 10e-6
mu = 1
R = np.linspace(0,40e-6,40)
growth_rate = mu*R/(k+R)
```

```python
plt.rcParams.update({'font.size': 14})  # increase the font size
plt.xlabel("t [s]")
plt.ylabel("R")
plt.plot(R, growth_rate, marker=".")
plt.title("Plot of the Monod function")
```

The differential equation (number(4) in the latex file) is solved numerically
\
\
$$\frac{dN}{dt} = \gamma \mu_{max} - N ( c \gamma k_{s}  + d)$$ 

```python
from scipy.integrate import odeint

def dN_dt(N, t):
    return g*mu-N*(c*g*k+d)

g = 5 # gamma=w*b , w = 0.5 <1 and b = 10
c = 1 # constant that depends on the species 
d = 0.01 # death rate
t = np.linspace(0,700,700)
N0 = 10  # the initial condition (initial population)
N = odeint(dN_dt, N0, t)
N = np.array(N).flatten()
```

```python
plt.rcParams.update({'font.size': 14})  # increase the font size
plt.xlabel("t [s]")
plt.ylabel("N")
plt.plot(t, N)
```

The numerical solution is compared with the exact one (numer (11) )
\
\
$N(t) = \left(N_{0} - \frac{\gamma \mu_{max}}{ c \gamma k_{s} +d} \right) e^{-t( c \gamma k_{s} +d)} + \frac{\gamma \mu_{max}}{ c \gamma k_{s} +d}$

```python
plt.rcParams.update({'font.size': 14})
N_exact = ( np.exp(-t*(g*c*k+d))*(N0-mu*g/(g*k*g+d)) ) + mu*g/(g*k*g+d)
#plt.plot(t, N, t, N_exact, "+")
plt.plot(t, N, color='b')
plt.plot(t, N_exact, color= 'm')
plt.xlabel("t")
plt.ylabel("N")
plt.legend(('numerical solution','exact solution'))
plt.plot(t, N)
```
